```python
import requests
Link=("https://www.cars24.com/blog/10-most-expensive-cars-in-the-world/")
Link_text=requests.get(Link).text
print(Link_text)
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(Link_text,'lxml')
print(soup)
```

```python
print(soup.title)
```

```python
def getdata(url):
    r = requests.get(url)
    return r.text
```

```python
htmldata = getdata("https://www.cars24.com/blog/10-most-expensive-cars-in-the-world/")
soup = BeautifulSoup(htmldata, 'html.parser')
for item in soup.find_all('img'):
    print(item)
```
